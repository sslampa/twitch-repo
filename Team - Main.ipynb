```python
all_teams_df = pd.concat([archon_df, c9_df, complexity_df, liquid_df, nihilum_df, tsm_df, tempostorm_df], ignore_index=True)
```

## 2. Aggregate Team Dataset

To be able to make the below plots the mean viewers (mean_viewers), mean followers (mean_followers), and number of streaming periods (count_views) are merged into the all_teams_agg dataframe by team and name. 

```python
mean_viewers = all_teams_df.groupby(by=['team', 'name'])['current_viewers'].mean()
mean_viewers = pd.DataFrame(mean_viewers).reset_index()
```

```python
mean_followers = all_teams_df.groupby(by=['team', 'name'])['followers'].mean()
mean_followers = pd.DataFrame(mean_followers).reset_index()
```

```python
count_views = all_teams_df.groupby(by=['team', 'name'])['current_viewers'].count()
count_views = pd.DataFrame(count_views).reset_index()
count_views.columns = ('team', 'name', 'count')
```

```python
new_df = pd.merge(mean_followers, mean_viewers, on=['team', 'name'])
```

```python
all_teams_agg = pd.merge(new_df, count_views, on=['team', 'name'])
```

## 3. Overview Of Teams Compared To Population

These are the same two plots found in the Main - Population page. The big difference between teams and the population lies in the tighter fit on the mean viewers vs followers plot. A possibility for this could be that streamers for teams are established and maintain a regular schedule. 

```python
sns.pairplot(all_teams_agg, x_vars=['count', 'followers'], y_vars=['current_viewers'], size=6, kind='reg')
```

## 4. Comparison Of Teams

These plots are the same as the above plots, but with an individual look at each team.

### Viewers Vs Followers

Each team shows the same general trend except for TSM. TSM only contains two members, so the curve given should be taken lightly. However, both streamers of TSM have a much higher follower count and that may also be a reason for their difference.

```python
colors = ['#e31837', '#15317e', '#724c9f', '#000000', '#787775', '#f2552c', '#60223b']
g = sns.lmplot('followers', 'current_viewers', data=all_teams_agg, hue='team', ci=None, size=8, scatter_kws={'s':100},
               palette=colors)
g.set(ylim=(0,22500))
```

### Viewers Vs Count

Once again TSM is the most notably different team; however, Tempo Storm and Complexity also show a negative curve. Based on this it seems there is no real correlation between how much exposure a streamer has and the amount of viewers those teams receive.

```python
g1 = sns.lmplot('count', 'current_viewers', data=all_teams_agg, hue='team', ci=None, size=8, palette=colors, scatter_kws={'s':100})
g1.set(ylim=(0,22500))
```

### Histogram Of Hours Each Team Streams

The histogram shows how many periods each team streams at the given hour. Each time recorded is based on military time and PDT time zone. For most teams, there is a noticeable downward trend starting ~8:00 and ending at ~15:00. It was shown in the Main - Population that this is prime hours for tournament streams, which a lot of these streamers participate in. The only team showing the opposite trend is Tempo Storm. The numbers may be skewed because these teams sometimes host their own tournaments.  

```python
all_teams_df.hist('hours', by='team', figsize=(12,8), bins=24, sharey=True, range=(0,23))
```

### Histogram Of Days Each Team Streams

The '0' denotes Monday and '6' denotes Sunday; each day then follows chronologically from there. For the most part, each team is consistent on which days they stream. C9 sees a dip towards the weekend, but those numbers are average compared to the rest of the teams. 

```python
all_teams_df['weekday'] = pd.DatetimeIndex(all_teams_df['current_date']).weekday
```

```python
all_teams_df.hist('weekday', by='team', figsize=(12,8), bins=7, sharey=True, range=(0,6))
```
