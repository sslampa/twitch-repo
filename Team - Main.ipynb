```python
import pandas as pd
import datetime
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import create_list
%matplotlib inline
```

```python
#archon
archon_df = pd.read_csv("archon.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
archon_df['team'] = 'archon'
```

```python
#c9
#amazhs, deernadia, hero_firebat, xixo, zalaehs, orange_hs
c9_df = pd.read_csv("c9.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
c9_df['team'] = 'c9'
```

```python
#complexity
complexity_df = pd.read_csv("complexity.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
complexity_df['team'] = 'complexity'
```

```python
#liquid
liquid_df = pd.read_csv("liquid.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
liquid_df['team'] = 'liquid'
```

```python
#nihilum
nihilum_df = pd.read_csv("nihilum.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
nihilum_df['team'] = 'nihilum'
```

```python
#tempostorm
tempostorm_df = pd.read_csv("tempostorm.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tempostorm_df['team'] = 'tempostorm'
```

```python
#tsm
tsm_df = pd.read_csv("tsm.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tsm_df['team'] = 'tsm'
```

```python
all_teams_df = pd.concat([archon_df, c9_df, complexity_df, liquid_df, nihilum_df, tsm_df, tempostorm_df], ignore_index=True)
```

```python
mean_viewers = all_teams_df.groupby(by=['team', 'name'])['current_viewers'].agg(np.mean)
mean_viewers = pd.DataFrame(mean_viewers).reset_index()
```

```python
mean_followers = all_teams_df.groupby(by=['team', 'name'])['followers'].agg(np.mean)
mean_followers = pd.DataFrame(mean_followers).reset_index()
```

```python
count_views = all_teams_df.groupby(by=['team', 'name'])['current_viewers'].count()
count_views = pd.DataFrame(count_views).reset_index()
count_views.columns = ('team', 'name', 'count')
```

```python
new_df = pd.merge(mean_followers, mean_viewers, on=['team', 'name'])
```

```python
new_df1 = pd.merge(new_df, count_views, on=['team', 'name'])
```

```python
sns.jointplot('count', 'current_viewers', data=new_df1, kind='reg', size=8)
```

```python
sns.jointplot('followers', 'current_viewers', data=new_df1, kind='reg', size=8)
```

```python
sns.set_style('whitegrid')
colors = ['#e31837', '#15317e', '#724c9f', '#000000', '#787775', '#f2552c', '#60223b']
g = sns.lmplot('followers', 'current_viewers', data=new_df1, hue='team', ci=None, size=10, scatter_kws={'s':100},
               palette=colors)
g.set(ylim=(0,22500))
```

```python
sns.set_style('whitegrid')
g1 = sns.lmplot('count', 'current_viewers', data=new_df1, hue='team', ci=None, size=10, palette=colors, scatter_kws={'s':100})
g1.set(ylim=(0,22500))
```
