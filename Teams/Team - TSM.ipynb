```python
import pandas as pd
import datetime
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import create_list
%matplotlib inline
```

## 2. Create Dataframe

```python
tsm_df = pd.read_csv("tsm.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tsm_df['current_date'] = pd.to_datetime(tsm_df['current_date'])
```

```python
datetime_df = create_list.make_empty_dataframe(tsm_df)
```

```python
tsm_df_agg = tsm_df.groupby('current_date').agg({'current_viewers': 'sum', 'followers': 'sum'})
tsm_df_agg.reset_index(inplace=True)
```

```python
temp_df = pd.merge(datetime_df, tsm_df_agg, how='outer', left_index=True, right_on='current_date')
temp_df.reset_index(inplace=True)
```

```python
temp_df['current_viewers'].fillna(0, inplace=True)
temp_df['week'] = pd.DatetimeIndex(temp_df['current_date']).week
temp_df['weekday'] = pd.DatetimeIndex(temp_df['current_date']).weekday
```

```python
temp_df.drop(['index', 'A'], 1, inplace=True)
temp_df.set_index('current_date', inplace=True)
temp_df.index.name = None
```

## 3. Team Specifics

### Average Viewers Vs Time

```python
fig2, axes= plt.subplots(nrows=1, ncols=4)
temp_df[(temp_df['week']==22) | (temp_df['week']==23)].plot(y='current_viewers', ax=axes[0], title='Week 1', figsize=(16,8), sharey=True, ylim=(0,35000))
temp_df[temp_df['week']==24].plot(y='current_viewers', ax=axes[1], title='Week 2')
temp_df[temp_df['week']==25].plot(y='current_viewers', ax=axes[2], title='Week 3', figsize=(20,4))
temp_df[(temp_df['week']==26) | (temp_df['week']==27)].plot(y='current_viewers', ax=axes[3], title='Week 4')
```
