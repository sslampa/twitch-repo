```python
import pandas as pd
import datetime
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import create_list
%matplotlib inline
```

## 1. Merging Team Datasets 

The making of the team datasets can be found in the 'Code - Data Cleaning' section. 

```python
df = pd.read_csv("june_final.csv", header=None, names=['id', 'name', 'followers', 'current_viewers', 'date_time'])
```

```python
archon_df = pd.read_csv("archon.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
c9_df = pd.read_csv("c9.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
complexity_df = pd.read_csv("complexity.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
liquid_df = pd.read_csv("liquid.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
nihilum_df = pd.read_csv("nihilum.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tempostorm_df = pd.read_csv("tempostorm.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tsm_df = pd.read_csv("tsm.csv", header=0, names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
```

```python
archon_df['team'] = 'archon'
c9_df['team'] = 'c9'
complexity_df['team'] = 'complexity'
liquid_df['team'] = 'liquid'
nihilum_df['team'] = 'nihilum'
tempostorm_df['team'] = 'tempostorm'
tsm_df['team'] = 'tsm'
```

```python
all_teams_df = pd.concat([archon_df, c9_df, complexity_df, liquid_df, nihilum_df, tsm_df, tempostorm_df], ignore_index=True)
```

## 2. Aggregate Team Dataset

To be able to make the below plots the mean viewers (mean_viewers), mean followers (mean_followers), and number of streaming periods (count) are merged into the all_teams_agg dataframe by team and name. 

```python
all_teams_agg = create_list.agg_df(all_teams_df, by=['team', 'name'])

# This dataframe contains all points
combined_streamer_final = create_list.agg_df(df, by=['name'])
```

## 3. Overview Of Teams Compared To Population

These are the same two plots found in the 'Main - Population' page, only with the teams as the red data points. The only discernible trend appears on the mean_viewers vs mean_followers plot. There appears to be a strong correlation between mean viewers and mean followers. A closer look at the plot shows a Pearson R value of .89 and p value of 5.9E-12, reasserting the strong correlation.

```python
ax = combined_streamer_final.plot('count', 'mean_viewers', kind='scatter', figsize=(10,8))
all_teams_agg.plot('count', 'mean_viewers', kind='scatter', ax=ax, color='red')
```

```python
ab =  combined_streamer_final.plot('mean_followers', 'mean_viewers', kind='scatter', figsize=(10,8))
all_teams_agg.plot('mean_followers', 'mean_viewers', kind='scatter', ax=ab, color='red')
```

```python
colors = ['#e31837', '#15317e', '#724c9f', '#000000', '#787775', '#f2552c', '#60223b']
g = sns.jointplot('mean_followers', 'mean_viewers', data=all_teams_agg, kind='reg', ci=None, size=8, scatter_kws={'s':80})
```

## 4. Comparison Of Teams

These plots are the same as the above plots, but with an individual look at each team.

### Viewers Vs Followers

Each team shows the same general trend except for TSM. TSM only contains two members, so the curve given should be taken lightly. However, both streamers of TSM have a much higher follower count and that may also be a reason for their difference.

```python
g = sns.lmplot('mean_followers', 'mean_viewers', data=all_teams_agg, hue='team', ci=None, size=8, palette=colors, scatter_kws={'s':100})
g.set(ylim=(0,22500))
```

### Viewers Vs Count

Once again TSM is the most notably different team; however, Tempo Storm and Complexity also show a negative curve. Based on this it seems there is no real correlation between how much exposure a streamer has and the amount of viewers those teams receive.

```python
g1 = sns.lmplot('count', 'mean_viewers', data=all_teams_agg, hue='team', ci=None, size=8, palette=colors, scatter_kws={'s':100})
g1.set(ylim=(0,22500))
```

### Histogram Of Hours Each Team Streams

The histogram shows how many periods each team streams at the given hour. Each time recorded is based on military time and PDT time zone. For most teams, there is a noticeable downward trend starting ~8:00 and ending at ~15:00 PDT. It was shown in the 'Main - Population' that this is prime hours for tournament streams, which a lot of these streamers participate in. The only team showing the opposite trend is Tempo Storm. The numbers may be skewed because these teams' players sometimes host their own tournaments.  

```python
all_teams_df.hist('hours', by='team', figsize=(12,8), bins=24, sharey=True, range=(0,23))
```

### Histogram Of Days Each Team Streams

The '0' denotes Monday and '6' denotes Sunday; each day then follows chronologically from there. For the most part, each team is consistent on which days they stream. C9 sees a dip towards the weekend, but those numbers are average compared to the rest of the teams. 

```python
all_teams_df['weekday'] = pd.DatetimeIndex(all_teams_df['current_date']).weekday
```

```python
all_teams_df.hist('weekday', by='team', figsize=(12,8), bins=7, sharey=True, range=(0,6))
```
