# The Best Fit

## 1. Summary

Here we set out to find if there are any current streamers that would support the popular Hearthstone teams. We needed to first create a dataset devoid of tournament streams and streamers already on a top team. From there, we created a minimum cutoff to ensure these streamers are able to help right away. This list was then cut further based on language and team restrictions. Each remaining streamer was then matched to a team.

```python
import pandas as pd
import datetime
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import create_list
%matplotlib inline
```

```python
# Reads the .csv data
main_df = pd.read_csv("main.csv")
no_tourn_df = pd.read_csv('no_tourn.csv')
main_df.drop('Unnamed: 0', 1, inplace=True)
no_tourn_df.drop('Unnamed: 0', 1, inplace=True)
```

```python
archon_df = pd.read_csv("archon.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
archon_df['current_date'] = pd.to_datetime(archon_df['current_date'])
c9_df = pd.read_csv("c9.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
c9_df['current_date'] = pd.to_datetime(c9_df['current_date'])
complexity_df = pd.read_csv("complexity.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
complexity_df['current_date'] = pd.to_datetime(complexity_df['current_date'])
liquid_df = pd.read_csv("liquid.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
liquid_df['current_date'] = pd.to_datetime(liquid_df['current_date'])
nihilum_df = pd.read_csv("nihilum.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
nihilum_df['current_date'] = pd.to_datetime(nihilum_df['current_date'])
tempostorm_df = pd.read_csv("tempostorm.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tempostorm_df['current_date'] = pd.to_datetime(tempostorm_df['current_date'])
tsm_df = pd.read_csv("tsm.csv", header=0, 
                 names=['name', 'followers', 'current_viewers', 'days', 'hours', 'minutes', 'current_date']).reset_index(drop=True)
tsm_df['current_date'] = pd.to_datetime(tsm_df['current_date'])
```

## 2. Created Aggregated Dataset

Standard subsetting/merging. Removed all tournament and streamers with a team.

```python
combined_team = pd.concat([archon_df, c9_df, complexity_df, liquid_df, nihilum_df, tempostorm_df, tsm_df])
combined_team_names = combined_team['name'].unique()
no_tourn_team_df = no_tourn_df[~no_tourn_df['name'].isin(combined_team_names)]
```

```python
tourn_count = pd.DataFrame(no_tourn_team_df.groupby('name')['followers'].count()).reset_index()
tourn_follo = pd.DataFrame(no_tourn_team_df.groupby('name')['followers'].mean()).reset_index()
tourn_views = pd.DataFrame(no_tourn_team_df.groupby('name')['current_viewers'].mean()).reset_index()
no_tourn_agg = pd.merge(tourn_count, tourn_follo, on='name')
no_tourn_agg = pd.merge(no_tourn_agg, tourn_views, on='name')
no_tourn_agg.columns = ('name', 'count', 'mean_followers', 'mean_viewers')
```

## 3. Determine Cutoff For Qualified Streamers

When determining the cutoff for qualified streamers, I decided to use the numbers at the first quartile for all teams (these numbers can be found on the 'Team - Main' page). Those numbers were chosen as it allows for streamers with lower, but still established numbers. As shown in the 'Single - Riser' page, when a person is added to a team their numbers can improve significantly. The following plots show a visual for how many streamers were removed based on those values. Each red data point represents the streamers that met both requirements.

```python
qualified_streamer = no_tourn_agg[(no_tourn_agg['mean_followers'] >= 16800) & (no_tourn_agg['count'] >= 186) & 
                                  (no_tourn_agg['mean_viewers'] >= 760)]
```

### Average Followers Vs Number Of Periods Streaming

```python
g = no_tourn_agg.plot('count', 'mean_followers', kind='scatter', figsize=(10,8))
qualified_streamer.plot('count', 'mean_followers', kind='scatter', ax=g, color='red')
plt.axhline(16800)
plt.axvline(186)
```

### Average Followers Vs Average Viewers

```python
h = no_tourn_agg.plot('mean_viewers', 'mean_followers', kind='scatter', figsize=(10,8))
qualified_streamer.plot('mean_viewers', 'mean_followers', kind='scatter', ax=h, color='red')
plt.axvline(760)
plt.axhline(16800)
```

## 4. Individual Look For Qualified Streamers

When choosing teams for each streamer, we tried to first match based on the hours the streamer streams then the days. Interestingly enough, the individual streamers streamed around 10:00 PDT where most teams saw lower numbers. This may be due to the streamers on the teams having to play in tournaments (check back to 'Twitch - Main' page to see how tournaments impacted streaming times).
